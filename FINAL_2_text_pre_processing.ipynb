## Clean up ,  Lemmatize, Remove Stopwords

```python
# Removing punctuation, empty spaces, lower instead of capital, remove numbers
def clean_up_and_token_lemmatizer(s):
    out = s.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'http\S+', '', out).lower()
    text1 = re.sub(r'\W', ' ', text)
    text2 = re.sub('\d*', '', text1).lstrip().rstrip()
    tokenized_words=word_tokenize(text2)
    return tokenized_words
    porterstemmer=PorterStemmer()
    #lemm=WordNetLemmatizer()

    tokenized_words=word_tokenize(text2)
    return tokenized_words
#     porterstemmer=PorterStemmer()
    lemm=WordNetLemmatizer()
    
    liste_lemm=[lemm.lemmatize(word) for word in tokenized_words]
#     liste_porter=[porterstemmer.stem(word) for word in tokenized_words]      
    return liste_lemm

def remove_stopwords(l):
    stop_words_english= get_stop_words('english')
    stop_words_spanish = get_stop_words('spanish')
    stop_words_english.extend(['ve', 'aa', 'aaa', 'from', 'subject', 're', 'phone', 'phones', 'cellphone', 'cellphones', 't', 'it', 'yoh', 'yey', 'just', 'wa'])
    
    # removing spanish stopwords
    filtered_words_sp = [word for word in l if word not in stop_words_spanish]
    filtered_words = [word for word in filtered_words_sp if word not in stop_words_english]
    return filtered_words

# the output is the serie with list of tokens
```

```python
t0 = time.time()
amazon2017_unpickled['text_processed']=amazon2017_unpickled.text.apply(lambda x: remove_stopwords(clean_up_and_token_lemmatizer(x)))
print("Done in %0.3fs." % (time.time() - t0))
```
